```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
years = [1960,1970 , 1980 , 1990 , 2000 , 2010 , 2020]
iranpop = [ 21.19 , 28.51 , 38.67 , 56.23 , 66.13 , 74.75 , 80.29]
plt.plot(years , iranpop)
plt.show()
```

```python
plt.scatter(years , iranpop)
plt.show()
```

```python
plt.hist(iranpop , bins ='auto')
plt.show()
```

```python
years = [1960,1970 , 1980 , 1990 , 2000 , 2010 , 2020]
iranpop = [ 21.19 , 28.51 , 38.67 , 56.23 , 66.13 , 74.75 , 80.29]
plt.figure(figsize = (11,8) , dpi=80)
plt.plot(years , iranpop)
plt.xlabel("Years")
plt.ylabel("Iran population")
plt.yticks([ 21.19 , 28.51 , 38.67 , 56.23 , 66.13 , 74.75 , 80.29] , ["21m" , '28m' , '38m' , '56m' , '66m' , '74m' , '80m'])
plt.show()
```

```python
plt.scatter(np.arange(7) , iranpop)
plt.xticks([0,1,2,3,4,5,6] , [1960,1970 , 1980 , 1990 , 2000 , 2010 , 2020])
plt.title("IRAN population")
plt.show()
```

```python
popsize = np.array([ 21.19 , 28.51 , 38.67 , 56.23 , 66.13 , 74.75 , 80.29])*2 
colors = ['violet' , 'red' , 'green' , 'blue' , 'yellow' , 'black' , 'grey']
plt.scatter(np.arange(7) , iranpop , s=popsize , c = colors )
plt.xticks([0,1,2,3,4,5,6] , [1960,1970 , 1980 , 1990 , 2000 , 2010 , 2020])
plt.margins(0.1)
plt.title("IRAN population")
plt.text(5.8 , 75 , "now" , fontsize=10)
plt.show()
```

```python
years = [1960,1970 , 1980 , 1990 , 2000 , 2010 , 2020]
iranpop = [ 21.19 , 28.51 , 38.67 , 56.23 , 66.13 , 74.75 , 80.29]
turkeypop = [ 25.19 , 29.51 , 34.67 , 52.23 , 61.13 , 78.75 , 82.29]
plt.plot(years , iranpop , ls = '-' , marker='+' , mew = 8)
plt.plot(years , turkeypop , ls = '--' , lw = 1 , marker='+' , mew = 8)
plt.legend(["iran" , 'turkey'] , loc="best")
plt.grid()
plt.show()
```

```python
plt.annotate("iran in present" , xytext=(2010,70) , xy = (2020,80) , arrowprops={"facecolor" : 'red' , 'width' :4 } ,fontsize=8)
plt.plot(years , iranpop , ls = '-' , marker='+' , mew = 8)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(years , iranpop , ls = '-' , marker='+' , mew = 8)
plt.title("IRAN")
plt.subplot(1,2,2)
plt.plot(years , turkeypop , ls = '--' , marker='+' , mew = 8)
plt.title("TURKEY")
plt.show()
```
